```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('sumanth.csv')
#data = pd.read_excel('Left_turn_90deg #1_straight.xlsx')
data
```

```python
pre_t = data['time']
```

```python
pre_t
```

```python
type(pre_t[1][-3:])
```

```python
# type(int(t[1][-3:]))
```

```python
lt = len(pre_t)
t = [0]

for i in range(1, lt):
  if int(pre_t[i][-3:]) > int(pre_t[i-1][-3:]):
    x = abs((int(pre_t[i][-3:]) - int(pre_t[i-1][-3:])))/1000
    #t.append((x + t[i-1]))
  else:
    x = abs((int(pre_t[i][-3:]) + 1000 - int(pre_t[i-1][-3:])))/1000
  t.append(np.round((x + t[i-1]), 3))

t
```

```python
df = pd.DataFrame (t)

## save to xlsx file

filepath = 'my_excel_file.xlsx'

df.to_excel(filepath, index=False)
```

```python
ax = data['ax']
ay = data['ay']
az = data['az']
gx = data['wx']
gy = data['wy']
gz = data['wz']
```

```python
plt.figure(figsize=(15,5))

plt.plot(t, ax, label='Ax')
plt.plot(t, ay, label='Ay')
plt.plot(t, az, label='Az')

plt.legend()
```
